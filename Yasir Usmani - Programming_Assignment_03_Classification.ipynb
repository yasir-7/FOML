<dt> <b> b. Implementation of Expectation and Maximization (EM) algorithm : </b> This part focuses on implementing the EM algorithm from scratch to estimate the parameters of the Gaussian Mixture Model (GMM). </dt>  

<dt>  1. Chose the number of mixtures (M) and initialize the parameters of GMM model  </dt>
<dd>  - Make sure that the diagonal elements of covariance are positive, and the weights of Gaussians are positive and sum to 1 for satisfying the axioms of probability.  </dd>

<dt>   2. Expectation step :  Evaluate the responsibility of Gaussians in generating the data points using the current estimated parameters i.e $\hat\mu_{k}$ , $\hat\Sigma_{k}$ and $\hat\pi_{k}$, here k indicates the Gaussian in GMM  </dt>
<br>

<dd>  <center> $\gamma_{nk} = \frac{\pi_{k} N ( \textbf{x}_{n} / \mu_{k} , \Sigma_{k} )}{ \sum_{m=1}^{M} \pi_{m} N ( x_{n} / \mu_{m} , \Sigma_{m} ) }    \;\;\;\;\;\; âˆ€ \;\;k=1,2,...,M \;\; and \;\; n=1,2,...,N$  </center>  </dd>

<br>

<dd>   Where N is the total number of data points and M is the total number of Gaussians in GMM   </dd>

<dd>   - $\gamma_{nk}$ represents the responsibility of $k^{th}$ Gaussian in generating the $n^{th}$ data point   </dd>

<dd>   - $\sum_{n=1}^{N}$ $\gamma_{nk}$ represents the the effective number of data points generated by $k^{th}$ Gaussian   </dd>


<dt>   3. Maximization step :  Update the parameters ($\hat\mu_{k}$ , $\hat\Sigma_{k}$ and $\hat\pi_{k}$) of GMM using the current resosibilities ($\gamma_{nk}$) by maximizing the likelihood. </dt>

<dd>
 - Effective number of data points generated by $k^{th}$ Gaussian : $N_{k}$ = $\sum_{n=1}^{N}$ $\gamma_{nk}$   
 </dd>

<dd>
 - Mean of $k^{th}$ Gaussian : $\hat{\mu_{k}} = \frac{1}{N_{k}} \sum_{n=1}^{N}\gamma_{nk} x_{n}$   
 </dd>

<dd>
 - Covariance of $k^{th}$ Gaussian : $\hat{\Sigma_{k}} = \frac{1}{N_{k}} \sum_{n=1}^{N} \gamma_{nk} ( x_{n} - \hat{\mu_{k}} )( x_{n} - \hat{\mu_{k}} )^{T}$   
 </dd>

<dd>
 - Weight of $k^{th}$ Gaussian : $\hat{\pi_{k}} = \frac{N_{k}}{N}$  
 </dd>

<dt>  4. Compute the log-likelihood with the updated parameters  </dt>
<dt>   5. Repeat the expectation and maximization steps untill convergence  </dt>

```python
########################################
# Part-2)-(a): Data generation
########################################

# Set the parameters of class distribution
mean1 = [1, 1]  # Mean of class1
mean2 = [2, 2]  # Mean of class2

cov = [[0.3, 0], [0, 0.3]]

# Generate training data. You can use "np.random.multivariate_normal()" function to sample data points from multivariate Gaussian distribution
class1_tr = np.random.multivariate_normal(
    mean1, cov, 50)  # Class1 training data
class2_tr = np.random.multivariate_normal(
    mean2, cov, 50)  # Class2 training data
# Combine class1 training and class2 training data to get total training data
tr_data = np.concatenate((class1_tr, class2_tr), axis=0)
# Class labels. Assign the value "0" to class1 and the value "1" to class2.
tr_targets = np.concatenate((np.zeros(50), np.ones(50)))

# Generate testing data.
class1_te = np.random.multivariate_normal(
    mean1, cov, 200)  # Class1 testing data
class2_te = np.random.multivariate_normal(
    mean2, cov, 200)  # Class2 testing data
# Combine class1 testing and class2 testing data to get total testing data
te_data = np.concatenate((class1_te, class2_te), axis=0)
# Class labels. Assign the value "0" to class1 and the value "1" to class2.
te_targets = np.concatenate((np.zeros(200), np.ones(200)))

########################################
# Part-2)-(b)
########################################
# Complete the below function
def LDA_classify(X_train, Y_train, X_test):
    # Inputs: Training data, Training labels, and Testing data
    # Outputs: Between class covariance

    # Compute the weight vector
    m1 = np.mean(X_train[Y_train == 1], axis=0)
    m2 = np.mean(X_train[Y_train == 0], axis=0)
    w =  m1 - m2
    w = w/np.linalg.norm(w)

    # Compute the threshold
    threshold = np.dot(w, (m1 + m2) / 2)

    # Project the data onto 1 dimensional space. Compare it with the threshold to make the decision
    projections = np.dot(X_test, w)
    predictions = (projections > threshold).astype(int)

    # Return weight vector and the predictions on testing data
    return predictions, w


predictions, w = LDA_classify(tr_data, tr_targets, te_data)


def LDA_project(w, X_test, Y_test):
    # Inputs: weight vector, testing data, and testing labels

    # Project the test data onto 1-dimensional space
    projections = (X_test @ w)

    # Create a line along the weight vector for visualization
    line = np.linspace(min(X_test[:, 0])-1, max(X_test[:, 0])+1, 100)

    # Compute y coordinates of projected points on the line
    slope = w[1] / w[0]
    intercept = -10
    y_line = slope * line + intercept

    # Plot the original data points
    plt.figure(figsize=(15, 5))
    plt.scatter(X_test[Y_test == 0, 0], X_test[Y_test ==
                0, 1], label='Class 0 - Original')
    plt.scatter(X_test[Y_test == 1, 0], X_test[Y_test ==
                1, 1], label='Class 1 - Original')

    # Plotting projected points on projection line
    y_proj = slope * projections + intercept
    plt.scatter(projections[Y_test == 0], y_proj[Y_test == 0], label='Class 0 - Projected')
    plt.scatter(projections[Y_test == 1], y_proj[Y_test == 1], label='Class 1 - Projected')

    # Plot the line along the weight vector
    plt.plot(line, y_line, color='brown', label='Projection Line')

    plt.legend()
    plt.show()


LDA_project(w, te_data, te_targets)

# Compute and print the accuracy
acc = np.mean(predictions == te_targets)
print('Accuracy with LDA:', acc*100)

# ########################################
# #Part-2)-(c)
# ########################################


def FDA_classify(X_train, Y_train, X_test):
    # Inputs: Training data, Training labels, and Testing data
    # Outputs: Weight vector and predictions

    # Compute the weight vector
    m1 = np.mean(X_train[Y_train == 1], axis=0)
    m2 = np.mean(X_train[Y_train == 0], axis=0)
    mean_diff = m1 - m2

    within_class_scatter = np.cov(
        X_train[Y_train == 0], rowvar=False) + np.cov(X_train[Y_train == 1], rowvar=False)
    w = np.linalg.inv(within_class_scatter).dot(mean_diff)
    w = w/np.linalg.norm(w)

    # Compute the threshold
    threshold = np.dot(w, (np.mean(
        X_train[Y_train == 1], axis=0) + np.mean(X_train[Y_train == 0], axis=0)) / 2)

    # Project the data onto 1 dimensional space. Compare it with the threshold to make the decision
    projections = np.dot(X_test, w)
    predictions = (projections > threshold).astype(int)

    # Return weight vector and the predictions on testing data
    return w, predictions


w, predictions = FDA_classify(tr_data, tr_targets, te_data)


def FDA_project(w, X_test, Y_test):
    # Inputs: weight vector, testing data, and testing labels

    # Project the test data onto 1-dimensional space
    projections = (X_test @ w)

    # Create a line along the weight vector for visualization
    line = np.linspace(min(X_test[:, 0])-1, max(X_test[:, 0])+1, 100)

    # Compute y coordinates of projected points on the line
    slope = w[1] / w[0]
    intercept = -10
    y_line = slope * line + intercept

    # Plot the original data points
    plt.figure(figsize=(15, 5))
    plt.scatter(X_test[Y_test == 0, 0], X_test[Y_test ==
                0, 1], label='Class 0 - Original')
    plt.scatter(X_test[Y_test == 1, 0], X_test[Y_test ==
                1, 1], label='Class 1 - Original')

    # Plotting projected points on projection line
    y_proj = slope * projections + intercept
    plt.scatter(projections[Y_test == 0], y_proj[Y_test == 0], label='Class 0 - Projected')
    plt.scatter(projections[Y_test == 1], y_proj[Y_test == 1], label='Class 1 - Projected')

    # Plot the line along the weight vector
    plt.plot(line, y_line, color='brown', label='Projection Line')

    plt.legend()
    plt.show()


FDA_project(w, te_data, te_targets)

# Compute and print the accuracy
acc = np.mean(predictions == te_targets)
print('Accuracy with FDA:', acc*100)

# ########################################
# #Part-2)-(d)
# ########################################


def FDA_project_K(X_train, Y_train, X_test, k):
    # Inputs: Training data, Training label, Testing data, and the number of eigen vectors to be considered
    # Oupur: Projection matrix

    # Compute the mean vectors for each class
    mean1 = np.mean(X_train[Y_train == 0], axis=0)
    mean2 = np.mean(X_train[Y_train == 1], axis=0)

    # Compute the within class scatter matrix
    S_W = np.zeros((X_train.shape[1], X_train.shape[1]))
    for i in range(X_train.shape[0]):
        row, mv = X_train[i].reshape(X_train.shape[1], 1), mean1.reshape(
            X_train.shape[1], 1) if Y_train[i] == 0 else mean2.reshape(X_train.shape[1], 1)
        S_W += (row-mv).dot((row-mv).T)

    # Compute the between class scatter matrix
    mean_diff = (mean1 - mean2).reshape(X_train.shape[1], 1)
    S_B = mean_diff.dot(mean_diff.T)

    eig_vals, eig_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))

    # Select the top k eigenvectors
    eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:, i]) for i in range(len(eig_vals))]
    eig_pairs = sorted(eig_pairs, key=lambda k: k[0], reverse=True)
    w = np.hstack([eig_pairs[i][1].reshape(X_train.shape[1], 1)
                  for i in range(k)])

    w = w/np.linalg.norm(w)

    # Return the projection matrix
    return w
```

```python
##################################################
#Part-5)-1: Data generation
##################################################

#Set the parameters of class distribution
mean1 = np.array([1, 1])  # Mean of class1
mean2 = np.array([2, 2])  # Mean of class2
cov = np.array([[0.3, 0.0], [0.0, 0.3]])  # Covariance matrix. Same covariance for class1 and class2

#Generate training data. You can use "np.random.multivariate_normal()" function to sample data points from multivariate Gaussian distribution
class1_tr = np.random.multivariate_normal(mean1, cov, 50)  # Class1 training data
class2_tr = np.random.multivariate_normal(mean2, cov, 50)  # Class2 training data
tr_data = np.concatenate((class1_tr, class2_tr), axis=0)  # Combine class1 training and class2 training data to get total training data
tr_targets = np.concatenate((np.zeros(50), np.ones(50)))  # Class labels. Assign the value "0" to class1 and the value "1" to class2.

#Generate testing data.
class1_te = np.random.multivariate_normal(mean1, cov, 200)  # Class1 testing data
class2_te = np.random.multivariate_normal(mean2, cov, 200)  # Class2 testing data
te_data = np.concatenate((class1_te, class2_te), axis=0)  # Combine class1 testing and class2 testing data to get total testing data
te_targets = np.concatenate((np.zeros(200), np.ones(200)))  # Class labels. Assign the value "0" to class1 and the value "1" to class2.

##################################################
#Part-5)-b: Complete the below logistic regression function
##################################################
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def Logistic_Regression(X_train, Y_train, X_test):
    # Inputs: Training data, Training labels, and Testing data
    # Outputs: Predictions on test data

    max_iter = 1000  # max number of iterations for parameter update
    threshold = 0.5  # threshold for classification
    X = np.hstack((np.ones((X_train.shape[0], 1)), X_train))  # Data. Append ones to the training data to take care of the bias
    w = np.random.rand(X.shape[1])  # Initialize the weight vector

    for iter in range(max_iter):
        preds = sigmoid(np.dot(X, w))  # Predictions on X. Sigmoid(w^T.x)
        error = preds - Y_train  # Difference between preds and Y_train

        gradient = np.dot(X.T, error)
        hessian = np.dot(np.dot(X.T, np.diag(preds * (1 - preds))), X)

        w -= np.dot(np.linalg.pinv(hessian), gradient)

    X_test = np.hstack((np.ones((X_test.shape[0], 1)), X_test))
    test_preds = sigmoid(np.dot(X_test, w))

    test_preds = (test_preds > threshold).astype(int)



    return w, test_preds

w, Y_pred = Logistic_Regression(tr_data, tr_targets, te_data)

###############################################
#Part-5)-c:
###############################################
def LR_Accuracy(Y_test, Y_pred):
    # Inputs: Ground truth and predicted labels of test data
    # Outputs: Accuracy

    # Compute and return the accuracy
    acc = np.mean(Y_test == Y_pred)
    return acc

acc = LR_Accuracy(te_targets, Y_pred)
print('\n\n Accuracy with Logistic Regression:', acc*100, '\n\n')

##########################################
#Part-5)-d: Plot the decision regions. Overlay the test points on the plot
##########################################

def plot_decision_boundary(X_test, Y_test, w):
    # Plot the points
    plt.scatter(X_test[Y_test == 0][:, 0], X_test[Y_test == 0][:, 1], label='Test Class 1', color='lightblue', edgecolor='blue')
    plt.scatter(X_test[Y_test == 1][:, 0], X_test[Y_test == 1][:, 1], label='Test Class 2', color='pink', edgecolor='red')

    # Plot the decision boundary line
    x1_min, x1_max =np.min(X_test[:, 0]), np.max(X_test[:, 0])
    xx1 = np.linspace(x1_min, x1_max, 100)
    xx2 = -(w[0] + w[1] * xx1) / w[2]  # Solve for x2 when w^Tx = 0
    plt.plot(xx1, xx2, color='purple', linestyle='--', linewidth=2, label='Decision Boundary')

    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()

plt.title('Decision Boundary')
plot_decision_boundary(te_data, te_targets, w)

##########################################
#Part-5)-e: Logistic regression for the data with true outliers
##########################################

outliers = np.array([[8, 5], [7, 6], [8, 9]])
class1_tr_outliers = np.concatenate((class1_tr, outliers), axis=0) # Add outliers to class 1 training data

# Combine class 1 with outliers and class 2 data
tr_data_outliers = np.concatenate((class1_tr_outliers, class2_tr), axis=0)
tr_targets_outliers = np.concatenate((np.zeros(53), np.ones(50)))

w, predictions = Logistic_Regression(tr_data_outliers, tr_targets_outliers, te_data)

# Plot the decision boundary/regions
plt.title('Logistic Regression - True Outliers')
plot_decision_boundary(te_data, te_targets, w)
##########################################
# Implement least squares method
##########################################

w, Y_pred = LS_Classify(tr_data_outliers, tr_targets_outliers, te_data)
plt.title('Least squares - True Outliers')
plot_decision_boundary(te_data, te_targets, w)
```

```python
#All imports
import numpy as np
import pandas as pd
import random as rand
import matplotlib.pyplot as plt
from scipy.stats import norm, multivariate_normal
import sys
import math

##########################################
#Part-6)-b: Implimentation of Expectation-Maximization (EM) algorithm
##########################################


############################################################################
#Write a function to compute the log-likelihood of the data given the parameters
############################################################################
def likelihood(data, pi, mu, sigma, log_likelihood=False, likelihoods=False):
    # Inputs: data, parameters of GMM model
    # Outputs: likelihood

    likelihood_total = 0
    likelihoods = []
    for point in data:
        point_likelihood = 0
        for k in range(len(pi)):
            point_likelihood += pi[k] * multivariate_normal.pdf(point, mean=mu[k], cov=sigma[k])
        likelihoods.append(point_likelihood)
        likelihood_total += np.log(point_likelihood)

    if log_likelihood:
        return likelihood_total
    elif likelihoods:
        return likelihoods
    else:
        return np.exp(likelihood_total)

############################################################################
#Write a function for the expectation step
############################################################################

def expectation(data, pi, mu, sigma):
    # Inputs: data, parameters of GMM model
    # Outputs: gamma

    N = len(data)
    M = len(pi)
    gamma = np.zeros((N, M))

    # Calculate responsibilities for each data point and component
    for n in range(N):
        for k in range(M):
            numerator = pi[k] * multivariate_normal.pdf(data[n], mean=mu[k], cov=sigma[k])
            denominator = sum(pi[j] * multivariate_normal.pdf(data[n], mean=mu[j], cov=sigma[j]) for j in range(M))
            gamma[n, k] = numerator / denominator

    return gamma


############################################################################
#Write a function for the maximization step.
############################################################################
def maximization(data, gamma):
    # Inputs: data, gamma
    # Outputs: Updated parameters of GMM model

    N = len(data)
    Nk = np.sum(gamma, axis=0)

    # Update pi
    pi_new = Nk / N

    # Update mu
    mu_new = np.dot(gamma.T, data) / Nk[:, np.newaxis]

    # Update sigma
    sigma_new = []
    for k in range(len(pi_new)):
        diff = data - mu_new[k]
        sigma_k = np.dot((gamma[:, k] * diff.T), diff) / Nk[k]
        sigma_new.append(sigma_k)

    return pi_new, mu_new, sigma_new


#Complete the below function. It plots the contours
def plot_contours(data, labels, pi, mu, sigma, title):
    # Inputs: data, labels, parameters of GMM model, title
    # Outputs: None (plots the contours)

    # Set up figure
    plt.figure(figsize=(8, 6))

    # Generate meshgrid
    x_min, x_max = -3, 3
    y_min, y_max = -3, 3
    x, y = np.mgrid[x_min:x_max:0.01, y_min:y_max:0.01]
    pos = np.dstack((x, y))

    # Plot contours for each component
    for k in range(len(pi)):
        rv = multivariate_normal(mu[k], sigma[k])
        plt.contour(x, y, rv.pdf(pos), alpha=0.5)

    # Scatter plot data points with color-coded labels
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')

    # Set title and labels
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')

    # Show plot
    plt.show()


############################################################################
#Run the expectation and maximzation algorithm to estimate the parameters
############################################################################
data = d_full
# Labels
labels = labels_full
# Number of input examples
N = len(data)
# Data dimension
D = data.shape[1]
# Number of components in GMM
M = 2

# Initialize the prior probabilities
pi = np.array([0.5, 0.5])
# Initialize the mean vectors
mu = np.array([[0.0, 0.0], [1.0, 1.0]])
# Initialize the covariance matrices

# Compute full covariance matrices
sigma_full = [np.identity(D) for _ in range(M)]

# Compute trace of the full covariance matrix
trace_full = np.trace(sigma_full[0])

# Compute spherical covariance matrices
sigma_spherical = [np.eye(D) * trace_full] * M

# Compute diagonal covariance matrices
sigma_diagonal = [np.diag(np.diag(sigma_full[i])) for i in range(M)]

nsteps = 10  # Number of steps to run EM algorithm
for index in range(0, nsteps, 1):
    # Compute the likelihood
    ll = likelihood(data, pi, mu, sigma_full, log_likelihood=True)

    # Expectation step
    gamma = expectation(data, pi, mu, sigma_full)

    # Maximization step
    pi, mu, sigma_full = maximization(data, gamma)

    # Plot the contours for full covariance
    plot_contours(data, labels, pi, mu, sigma_full, f'Full Covariance - Iteration {index}')

# Define function to run EM algorithm for a specific covariance type
def run_em_algorithm(data, pi, mu, covariance_matrices, plot_title):
    nsteps = 10
    for index in range(0, nsteps, 1):
        # Compute the likelihood
        ll = likelihood(data, pi, mu, covariance_matrices, log_likelihood=True)

        # Expectation step
        gamma = expectation(data, pi, mu, covariance_matrices)

        # Maximization step
        pi, mu, covariance_matrices = maximization(data, gamma)

        # Plot the contours
        plot_contours(data, labels, pi, mu, covariance_matrices, plot_title.format(index))

# Re-initialize for spherical covariance
pi = np.array([0.5, 0.5])
mu = np.array([[0.0, 0.0], [1.0, 1.0]])
run_em_algorithm(data, pi, mu, sigma_spherical, 'Spherical Covariance - Iteration {}')

# Re-initialize for diagonal covariance
pi = np.array([0.5, 0.5])
mu = np.array([[0.0, 0.0], [1.0, 1.0]])
run_em_algorithm(data,pi, mu, sigma_diagonal, 'Diagonal Covariance - Iteration {}')
```

<b>Part-6-d: Understanding Issues with EM Algorithm</b>

1. Convergence Speed: The EM algorithm's rate of convergence is significantly affected by initialization. While poorly initialised data might cause convergence to lag or result in local maxima rather than the global maximum, well-chosen initial parameters can speed up convergence.
2. Fit Quality: The ultimate parameters acquired during EM iterations are significantly dependent on the starting parameter selections. The probability of the algorithm convergent to a model that fits the data well is higher when the method starts nearer the genuine parameters. On the other hand, improper initialization might lead to less-than-ideal fits and possibly suck the algorithm into local optima.
3. Illustration of EM Iterations: Notably, with better initialization, Gaussian component contours swiftly align with the data distribution, whereas poor initialization necessitates more drastic adjustments in early iterations as the algorithm endeavors to correct its trajectory.
4. Maximum versatility is provided by full covariance, which captures all variation and covariance between features. However, in order to accurately estimate parameters without overfitting, this flexibility needs additional data. In general, comprehensive covariance matrices produce better fits when dealing with complicated, correlated data.
5. Assumes feature independence with variance fluctuation among features in the context of diagonal covariance. Compared to complete covariance, this reduction allows for faster convergence and reduces the danger of overfitting, particularly in cases when feature independence is maintained. However, it may fail to take feature correlations into account, leading to less accurate fits.
6. Spherical covariance: Provides computing efficiency and is appropriate for basic data distributions; it assumes equal variance and feature independence. It is, however, rigid and may result in poor fits if data distributions depart from these presumptions.
7. An example of EM iterations Gaussian component contours with spherical covariance are circular and only vary in size. While complete covariance permits more complex ellipses that may precisely adjust to data distributions, diagonal covariance yields ellipses that are aligned with axes. The influence of these assumptions on the model's capacity to represent actual data distributions over iterations is made clear by visualising them.
